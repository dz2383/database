### STEP 1
Scrape the first page of judicial vacancies:

http://www.uscourts.gov/judges-judgeships/judicial-vacancies/current-judicial-vacancies

```python
# My libraries

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait

import pandas as pd
import time
import re
import numpy as np
```

```python
# Start to scraping

driver = webdriver.Chrome()
driver.get('http://www.uscourts.gov/judges-judgeships/judicial-vacancies/current-judicial-vacancies')

# This is all data of the webpage
all_info = driver.find_element_by_xpath('//*[@id="content"]/table/tbody')
# all_info.text

# create a list of dictionaries

vacancy_list=[]

for row in all_info.find_elements_by_tag_name('tr'):
    data={}
    data['court'] = row.find_elements_by_tag_name('td')[0].text
    data['incumbent']= row.find_elements_by_tag_name('td')[1].text
    data['vacancy_reason']= row.find_elements_by_tag_name('td')[2].text
    data['vacancy_date']= row.find_elements_by_tag_name('td')[3].text
    data['nominee']= row.find_elements_by_tag_name('td')[4].text
    data['nominee_date']= row.find_elements_by_tag_name('td')[5].text
    
    vacancy_list.append(data)
    
vacancy_list

# This is a dataframe of vacancy_list of judges
df=pd.DataFrame(vacancy_list)
df
```

```python
import json

#
# This will maybe make more sense if you go to geojson.io and look at the output
# 

# A function to convert each row into a GeoJSON Feature
# Right now it makes a LineString from all of the coordinates, but 
# you could reorder them or make them a Polygon (geometry type: Polygon)
# Everything in the row gets converted into a dictionary,
# and put into 'properties'

# Have to use 'eval' to turn the string "(45.44, -74.0)" into an actual list

def to_feature(row):
    return {
      "type": "Feature",
      "properties": dict(row),
      "geometry": {
        "type": "LineString",
        "coordinates": [
            list(reversed(eval(row['Birth_geo']))),
            list(reversed(eval(row['School1_geo']))),
            list(reversed(eval(row['School2_geo']))),
            list(reversed(eval(row['recent_career_geo']))),
            list(reversed(eval(row['current_geo'])))
        ]
      }
    }

# Use .apply to run to_feature on every row, save the result
features = df4.apply(to_feature, axis=1)

# Now build a collection of features (that's what GeoJSON is)
geojson = {
  "type": "FeatureCollection",
  "features": list(features)
}

# Write it to a file called output.json
# We use indent to make it look a little nicer in case
# we have to read it
with open("output_secondcircuit.js", 'w') as file:
    data = json.dumps(geojson, indent=4)
    file.write(data)
```
